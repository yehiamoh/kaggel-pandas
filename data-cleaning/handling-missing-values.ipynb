```python
import pandas as pd
import numpy as np


games=pd.read_csv('../data/NFL Play by Play 2009-2016 (v3).csv')
col=games.columns

missing_values=games.isnull().sum().sort_values(ascending=False) # get the each column value null data summation (it's type is a Series)
missing_values[0:10]

total_cells=np.prod(games.shape)
total_missing_values=games.isnull().sum().sum() #missing_values.sum()  first sum sums each column null values second sum sum all columns null values  (first sum return a series) (second sum sum all sereis values)
   
null_values_percentage=(total_missing_values/total_cells)*100
null_values_percentage # percenateg of null values in the dataframes


column_with_na_dropped=games.dropna(axis=1) # remove columns with at least one na value

# here you can see the differnce of the columns numbers after and befor removing
print(f'original DataFrame',games.shape)
print(f'After removing columns with na DataFrame',column_with_na_dropped.shape)

# replace all Na with a certain value you assign
games.fillna(0)

## replace all NA's the value that comes directly after it in the same column, 
# then replace all the remaining na's with 0

games.fillna(method="bfill",axis=0).fillna(0)
```
