```python
import pandas as pd
reviews=pd.read_csv("./data/ramen-ratings.csv")


reviews['Stars'] = pd.to_numeric(reviews['Stars'], errors='coerce') #convert the Stars column into number variable

reviews.groupby('Country').Stars.min() # Group the data by the Country and did and operation to the data related to each country by getting the minimun star each country got
reviews.groupby(['Country','Brand'])["Stars"].agg([min,max,len])

reviews.sort_values(by='Stars',ascending=False)
```
